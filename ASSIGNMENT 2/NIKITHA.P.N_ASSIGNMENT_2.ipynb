```python
import pandas as pd
import seaborn as s
import matplotlib.pyplot as plt
d=pd.read_csv("House Price India.csv")
plt.figure(figsize=(8,6))
print(d.head(5))
```

```python
s.histplot(d['living area'])
```

```python
s.histplot(d['living area'], bins=40, kde=True)
plt.title('Distribution of Living Area')
plt.xlabel('Living Area')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8,7))
s.histplot(d['Built Year'], bins=40)
plt.title('year wise build datas')
plt.xlabel('Years')
plt.ylabel('Number of Houses Built')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
s.scatterplot(data=d, x='living area', y='Price')
plt.title('Scatter Plot of Living Area vs. Price')
plt.xlabel('Living Area')
plt.ylabel('Price')
plt.show()
```

```python
correlation_matrix = d.corr()
plt.figure(figsize=(10, 8))
s.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix Heatmap')
plt.show()
```

Descriptive Statistics:


```python
d=pd.read_csv("House Price India.csv")
des=d.describe()
head = d.head()

# Display the data types and non-null values for each column
info = d.info()

# You can also calculate additional statistics like median, variance, etc. as needed.
median = d.median()
variance = d.var()

# Print the descriptive statistics
print("Basic Descriptive Statistics:")
print(des)
print("\nFirst Few Rows of the DataFrame:")
print(head)
print("\nData Types and Non-Null Values:")
print(info)
print("\nMedian Values:")
print(median)
print("\nVariance:")
print(variance)
```
